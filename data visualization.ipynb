```python
plt.figure(figsize=(15,6))
#creating countplot of passenger by class
sns.countplot(x='Pclass',data=df)
#should write Pclass only why??
#based on pclass column in the data
```

```python
#we got third class as the highest one
#if we want to change the obtained figure size then
plt.figure(figsize=(8,8))
#to set titles and axis labels we will use
plt.title('required title')
plt.xlabel('title')
plt.ylabel('title')
#using all these written below
#to show the figure we need to write
plt.show()
```

```python
plt.figure(figsize=(15,6))

sns.countplot(x='Pclass',data=df)
plt.title('titanic data')
plt.xlabel('class number')
plt.ylabel('counting')
plt.show()
#if plt.show() not written then some other data might also be produced
#names changed
```

```python
#similar to count plot we can also create violinplot which will produce sharp parabola like structures
plt.figure(figsize=(10,6))
#voilinplot of age by class 
#we can select y axis if we require here let us select y axis with age
#if not default is taken write same as the name in the column
sns.violinplot(x='Pclass',y='Age',data=df)
plt.title('Age distribution')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

```python
#inference here we can see the people with all ages are present in the first class some less in second
#to work on the correlation part we need to use heat map
#similar to statistics and probability we can observe positive corelation ,negative correlation which is less moreate high
#except diagnol both sides there is correlation
#we can be able to observe that when the passenger class is increasing then fare will be decreasing 1st class more cost 2nd class less
#so it indicates negative correlation
```

```python
#heat map
x=df.corr()
#obtaining correlation
plt.figure(figsize=(10,6))
sns.heatmap(x,annot=True)
#annot refers to annotation if we remove it will not show the values only plain graph shown
plt.title('correlation matrix')
plt.xlabel('features')
plt.ylabel('features')
plt.show()
```

```python
#using swarmplot to find the inference between passenger and fare
plt.figure(figsize=(10,6))
#creating swarmplot of fares by class
sns.swarmplot(x='Pclass',y='Fare',data=df)
#xaxis take categorical yaxis numerical
plt.title('fare by class')
plt.xlabel('class')
plt.ylabel('fare')
plt.show()
```

```python
#inference betwwen the passenger class and fare is obtained
#passenger class 1 more 2 second more 3rd last moderately negative correlation
```

```python
#inorer to study about the distribution of data of fare and Pclass then use displot we need to define one attribute with 
#hue so that we can find the difference betweeen the list of items present in it
#here we define hue with pclass x axis with fare y axis pclass according to the count and how data is distributed
#we will get the curves similar to frequency distribution curves in statistics and probability
```

```python
sns.distplot(df["Age"])
plt.show()
#used to show wheather the data is normally distributed or not- distribution plot
```

```python
#if we want only histogram and not the curve then kde=false should be asigned
sns.distplot(df["Age"],kde=False)
plt.show()
```
